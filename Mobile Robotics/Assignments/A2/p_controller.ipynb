```python
from scipy.integrate import odeint
import numpy as np
```

Part a :
P controller

kp = 5

```python
def model(u,t):
    h = u[0]
    h1 = u[1]
    c = 0.162 #kp = 5
    f = [[],[]]
    f[0] = h1
    f[1] = c - (c * h)
    return f
```

```python
time = np.linspace(0,50)
```

```python
sol = odeint(model,[0,0],time)
```

```python
import matplotlib.pyplot as plt
plt.plot(time,sol[:,0],'r')
plt.plot(time,sol[:,1],'b')
plt.legend(['h (ODEINT)','dh/dt (ODEINT)'])
plt.xlabel('Time')
plt.show()
```
